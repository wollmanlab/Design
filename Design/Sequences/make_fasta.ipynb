```python
from tqdm import tqdm
def revcomp(seq,model='infer'):
    if model=='infer':
        if 'U' in seq:
            model='rna'
        else:
            model='dna'
    if model=='dna':
        converter = {'A':'T','T':'A','C':'G','G':'C'}
    else:
        converter = {'A':'U','U':'A','C':'G','G':'C'}
    return ''.join(reversed([converter[i] for i in str(seq)]))
def fuzzy_search(seq, target, max_errors=1):
    """
    Search for a target string in a sequence with allowed errors.
    
    Args:
        seq (str): The sequence to search in (can contain ACGTU)
        target (str): The target string to search for (can contain ACGTU)
        max_errors (int): Maximum number of allowed errors (default: 1)
    
    Returns:
        bool: True if target is found with allowed errors, False otherwise
    """
    target_len = len(target)
    seq_len = len(seq)
    
    # Check if target is longer than sequence
    if target_len > seq_len:
        return False
    
    # Iterate through all possible starting positions
    for i in range(seq_len - target_len + 1):
        substring = seq[i:i + target_len]
        errors = 0
        
        # Count mismatches
        for j, (sub_char, target_char) in enumerate(zip(substring, target)):
            if sub_char != target_char:
                errors += 1
                # Early termination if we exceed max_errors
                if errors > max_errors:
                    break
        
        # If errors are within limit, return True (found)
        if errors <= max_errors:
            return True
    
    return False
bad_readout_sites = []
restriction_sites = []
site = 'GCGGCCGC' # NotI-HF
restriction_sites.append(site)
restriction_sites.append(revcomp(site))
for site in restriction_sites:
    for idx,row in Readout_sites.iterrows():
        seq = row['Sequence']
        if fuzzy_search(seq,site,max_errors=2):
            errors = 2
            if fuzzy_search(seq,site,max_errors=1):
                errors = 1
                if fuzzy_search(seq,site,max_errors=0):
                    errors = 0
                print(row['ID'],seq,site,idx,errors)
                bad_readout_sites.append(row['ID'])
available_readout_sites = [i for i in Readout_sites['ID'] if i not in bad_readout_sites]
```

```python
  def filter_csv_to_df(csv_file, gene_list):
    """
    Filters a CSV file to keep only rows where the gene column matches a list of genes
    and loads the filtered data into a pandas DataFrame.  Prioritizes speed by
    loading the entire CSV into memory first.

    Args:
        csv_file: Path to the input CSV file.
        gene_list: List of genes to keep.

    Returns:
        A pandas DataFrame containing the filtered rows.
    """

    # Load the entire CSV into a DataFrame
    df = pd.read_csv(csv_file)

    # Convert gene_list to a set for faster lookup (O(1) vs O(n))
    gene_set = set(gene_list)

    # Filter the DataFrame.  Use .isin() for efficient set membership testing.
    #  Also, access the gene column directly using its name, if known, for speed.
    #  We assume the gene column is named 'gene'.  If it's different, change this!
    try:
        filtered_df = df[df['gene'].isin(gene_set)]
    except KeyError:
        #handling the case that gene column not called gene
        df.columns = [str(col) for col in df.columns]   #make sure all the columns are string to be consistent with reading with csv files
        filtered_df = df[df.iloc[:, -4].isin(gene_set)] #use the original -4 column
    return filtered_df
    
def optimize_weight_matrix_to_fasta(used_WeightMat,available_probes):
    # Filter out zero-sum rows first
    nonzero_weights = used_WeightMat[used_WeightMat.sum(axis=1) > 0]
    available_probes = available_probes[available_probes['gname'].isin(nonzero_weights.index)]
    total_reduction = 0
    grouped_probes = dict(tuple(available_probes.groupby('gname')))

    all_rows = []
    for gene, row in tqdm(used_WeightMat.iterrows(), total=used_WeightMat.shape[0]):
        if row.sum() == 0:
            continue
            
        n_probes = int(row.sum())
        gene_probes = grouped_probes.get(gene)
        if gene_probes is None:
            continue

        # remove probes with restriction site match
        # restriction_sites = ['CACTGC','GCTAGC',revcomp('CACTGC'),revcomp('GCTAGC')]
        restriction_sites = []
        site = 'GCGGCCGC' # NotI-HF
        restriction_sites.append(site)
        restriction_sites.append(revcomp(site))
        # site = 'CCTGCAGG'
        # restriction_sites.append(site)
        # restriction_sites.append(revcomp(site))
        # site = 'GCTAGC'
        # restriction_sites.append(site)
        # restriction_sites.append(revcomp(site))

        # restriction_sites = ['GCTAGC',revcomp('GCTAGC')]
        for site in restriction_sites:
            gene_probes = gene_probes[[fuzzy_search(row['seq'],site,max_errors=1)==False for idx,row in gene_probes.iterrows()]]

        # Create bits array directly without string operations
        assigned_bits = np.repeat(np.arange(len(row)), row.astype(int))
        np.random.shuffle(assigned_bits)
        
        if n_probes > gene_probes.shape[0]:
            total_reduction += n_probes - gene_probes.shape[0]
            print(gene, n_probes, gene_probes.shape[0])
            assigned_bits = assigned_bits[:gene_probes.shape[0]]
        
        # Select random indices and create result in one step
        selected_indices = np.random.choice(gene_probes.index, len(assigned_bits), replace=False)
        selected_probes = gene_probes.loc[selected_indices].copy()
        selected_probes['bit'] = assigned_bits
        all_rows.append(selected_probes)
    print(f"Total reduction: {total_reduction}")
    result = pd.concat(all_rows)
    return result
csv_file = '/scratchdata1/ExternalData/mm10_probes_Oct28_2022.csv'
gene_list = E.index
fasta_probes = optimize_weight_matrix_to_fasta(E,filter_csv_to_df(csv_file, gene_list))
fasta_probes
```
